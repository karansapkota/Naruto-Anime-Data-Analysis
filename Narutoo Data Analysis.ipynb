```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('data/naruto_ratings_data_v02.csv')
```

```python
df1 = pd.read_csv("data/naruto_ratings_data_v03.csv")
```

```python
df2 = pd.read_csv("data/naruto_ratings_data_v04.csv")
```

```python
combined_df = pd.concat([df,df1,df2])
combined_df.head().T
```

```python
combined_df.drop(columns=['type', 'english_air_date'], inplace=True)
```

```python
combined_df.drop_duplicates(subset='Title', inplace=True)  # remove duplicate titles
```

# Data Visualization and Analysis

```python
# What is the average rating for Naruto Shippuden?
avg_rating = combined_df['rating'].mean()
print('The average rating for Naruto Shippuden is:', round(avg_rating, 2))
```

```python
# Total episodes in Naruto Shippuden ?
num_episodes = combined_df['episode_number_overall'].max()
print('Naruto Shippuden has', num_episodes, 'episodes.')
```

```python
# Find the top 10 highest rated episodes
top_10 = combined_df.nlargest(10, 'rating')
print("Top 10 Highest Rated Episodes:")
print(top_10[['Title', 'rating']])
```

```python
# Which episodes have the highest rating?
highest_rated = combined_df.loc[combined_df['rating'].idxmax()]
print('The highest rated episode is', highest_rated['Title'], 'with a rating of', highest_rated['rating'])
```

```python
# Find the top 10 lowest rated episodes
top_10 = combined_df.nsmallest(10, 'rating')
print("Top 10 Highest Rated Episodes:")
print(top_10[['Title', 'rating']])
```

```python
# Which episodes have the lowest rating?
lowest_rated = combined_df.loc[combined_df['rating'].idxmin()]
print('The lowest rated episode is', lowest_rated['Title'], 'with a rating of', lowest_rated['rating'])
```

```python
# What is the distribution of ratings for Naruto Shippuden?
sns.histplot(data=combined_df, x='rating', kde=True)
plt.title('Distribution of Ratings for Naruto Shippuden')
plt.xlabel('Rating')
plt.show()
```

We can see that the ratings of Naruto Shippuden follow a normal distribution with a peak around 7.5-8.0. The average rating of Naruto Shippuden is 8.18, which is quite good.

```python
# Plot the number of episodes per season
fig, ax = plt.subplots()
season_counts = combined_df['season'].value_counts().sort_index()
season_counts.plot(kind='bar', ax=ax)
ax.set_title('Number of Episodes per Season')
ax.set_xlabel('Season')
ax.set_ylabel('Number of Episodes')
plt.show()
```

```python
# Group the data by season and calculate the mean rating for each season
season_ratings = combined_df.groupby('season')['rating'].mean()
print("Mean Ratings by Season:")
print(season_ratings)
```

```python
# Group the data by season and calculate the mean rating for each season
season_ratings = combined_df.groupby('season')['rating'].mean()
# Create a bar plot to visualize the ratings for each season
season_ratings.plot(kind='bar')
plt.title('Season-wise Ratings')
plt.xlabel('Season')
plt.ylabel('Rating')
plt.show()
```

```python
# plot correlation between votes and rating
sns.scatterplot(x='votes', y='rating', data=combined_df)
plt.title('Correlation between Votes and Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.show()
```

```python
# Create a histogram to visualize the distribution of episode lengths
combined_df['episode_number_in_season'].plot(kind='hist', bins=20)
plt.title('Distribution of Episode Lengths')
plt.xlabel('Length (minutes)')
plt.show()
```
